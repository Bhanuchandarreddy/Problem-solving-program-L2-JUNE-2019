```python

# Graph to plot distance and time

distance = list(range(2,11,2))
time = list(range(1,6))
distance
#d=2*t
plt.title('Distance vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'ro')
plt.grid()
plt.show()
```

```python
t=list(range(1,11))
v=[2 * i for i in t]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i] * i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# numbers vs cubes
n=list(range(10))
sqn=[i**4 for i in n]
plt.plot(n,sqn)
plt.grid()
plt.show()
```

```python
# Voltage - Time
import numpy as np
t=[i/10 for i in range(1,21)]
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('voltage')
plt.show()
```

```python
# Voltage - Time
import numpy as np
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('voltage')
plt.show()
```

```python
# Voltage - Time
import numpy as np
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('voltage')
plt.show()
plt.savefig('images/voltage_Time.png')
```

```python
# income of all states from 2005 to 2013
df
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title('All states Income')
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid()
plt.show()
```
